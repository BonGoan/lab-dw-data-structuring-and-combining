```python
import pandas as pd

# Load the datasets
url1 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"
url2 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv"
url3 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv"

df1 = pd.read_csv(url1)
df2 = pd.read_csv(url2)
df3 = pd.read_csv(url3)

df1
```

```python
df2
```

```python
df3
```

```python
df1 = df1.rename(columns={'ST': 'State', 'GENDER' : 'Gender'})
df1
```

```python
df2 = df2.rename(columns={'ST': 'State' , 'GENDER' : 'Gender'})
df2
```

```python


# merging columns with duplicated
df2['State'] = df2['State'].combine_first(df2['State'])
df3['State'] = df3['State'].combine_first(df3['State'])

```

```python


# Similar handling for other overlapping columns
df2['Education'] = df2['Education'].combine_first(df2['Education'])
df3['Education'] = df3['Education'].combine_first(df3['Education'])
df2['Gender'] = df2['Gender'].combine_first(df2['Gender'])
df3['Gender'] = df3['Gender'].combine_first(df3['Gender'])

```

```python
df2['Customer Lifetime Value'] = df2['Customer Lifetime Value'].combine_first(df2['Customer Lifetime Value'])
df3['Customer Lifetime Value'] = df3['Customer Lifetime Value'].combine_first(df3['Customer Lifetime Value'])
```

```python
df_combined = pd.concat([df1, df2, df3], ignore_index=True)
df_combined = df_combined.drop_duplicates()

# Handling missing values (if needed)
df_combined = df_combined.fillna('Unknown') 

#reset index
df_combined = df_combined.reset_index(drop=True)

df_combined
```
